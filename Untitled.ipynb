```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mim
from scipy import spatial
from matplotlib import cm
from PIL import Image
```

```python
emoji_array = np.load('emojis_16.npy')
```

```python
emoji_mean = np.array([ar.mean(axis=(0,1)) for ar in emoji_array])
```

```python
tree = spatial.KDTree(emoji_mean)
```

```python
G_sm = np.array(Image.open('car.jpeg').resize([50,50]).getdata()).reshape([50,50,3])/256
```

```python
plt.figure()
plt.imshow(G_sm)
plt.title('Original Image')
```

```python
indices = []
flattened_img = G_sm.reshape(-1, G_sm.shape[-1]) 
```

```python
for pixel in flattened_img:
    pixel_ = np.concatenate((pixel, [1]))
    _, index=tree.query(pixel_)
    indices.append(index)
```

```python
emoji_matches = emoji_array[indices] 
#from index get the corresponding emoji (flattened)

dim = G_sm.shape[0] 
resized_ar = emoji_matches.reshape((dim, dim, 16, 16,4 ))
```

```python
final_img = np.block([[[x] for x in row] for row in resized_ar]) 
```

```python
plt.figure()
plt.imshow(final_img)
plt.title('Emoji mosaic')
plt.savefig('mosaic-img.png')
```
